```python
df_all_movies = pd.read_csv('/content/drive/MyDrive/archive_stores/databases-raw/movies.csv',sep='|')
df_all_movies['profissao'] = df_all_movies['profissao'].astype(str)
```

```python
# Removendo os dois primeiros caracteres "tt" da coluna de id dos filmes
df_all_movies['id'] = df_all_movies['id'].apply(lambda x: x[2:])
```

```python
df_all_movies['id'] = df_all_movies['id'].astype(int)
df_all_movies['id'].dtype
```

```python
# Obtendo o número de linhas e colunas
df_all_movies.shape
```

#### Para filtrar os dados de acordo com o recorte temporal escolhido para a análise (a partir de 1970), primeiro foi necessário transformar valores inconsistentes em valores nulos

```python
# Lista de todas as colunas do df
colunas = df_all_movies.columns

# Iterando sobre todas as colunas e aplicando a substituição por valor nulo
for coluna in colunas:
    df_all_movies[coluna] = df_all_movies[coluna].replace({"\\N": None, "\\n": None})
```

#### Para aplicar o filtro na coluna "anoLancamento" foi preciso converter os dados para inteiros

```python
# Substituindo os valores nulos na coluna 'anoLancamento' por 0 para converter o tipo de dado
df_all_movies['anoLancamento'].fillna(0, inplace=True)
```

```python
df_all_movies['anoLancamento'].dtype
```

```python
df_all_movies['anoLancamento'] = df_all_movies['anoLancamento'].astype(int)

df_all_movies['anoLancamento'].dtype
```

```python
# Encontrando os valores únicos na coluna 'anoLancamento'
valores_unicos = df_all_movies['anoLancamento'].unique()

# Exibindo os valores únicos
print(valores_unicos)
```

#### Os dados nulos da coluna 'genero' foram substituídos por string vazia. Isso porque o Pandas não permite fazer indexação direta de um DataFrame usando uma série booleana que contenha valores NA (NaN) no seu interior.

```python
# Preenchendo valores nulos na coluna 'genero' com uma string vazia
df_all_movies['genero'].fillna('', inplace=True)

# Filtro para incluir linhas que contenham "crime" na coluna 'genero'
filtrado_crime = df_all_movies['genero'].str.contains('crime', case=False)

# Aplicando o filtro
df_filtrado = df_all_movies[filtrado_crime]
```

```python
# Filtro para incluir apenas filmes a partir de 1970
df_filtrado = df_filtrado[df_filtrado['anoLancamento'] >= 1970]
```

```python
df_filtrado.rename(columns={"id": "idFilme"}, inplace=True)
```

```python
df_filtrado.head(7)
```

#### Um novo DataFrame contendo dados relacionados ao elenco foi gerado e salvo no formato CSV para possíveis análises posteriores

```python
df_elenco = df_filtrado[['idFilme','tituloPincipal','generoArtista','nomeArtista','profissao','titulosMaisConhecidos']]
```

```python
# salvando csv com dados de elenco
os.chdir('/content/drive/MyDrive/archive_stores/dados-trusted')
df_elenco.to_csv('df_elenco.csv', index=False)
```

### Continuando o preparo dos dados dos filmes...

#### Colunas que não são necessárias às análises pretendidas foram removidas do dataframe

```python
# Lista de nomes das colunas para remover
colunas_a_remover = ['tituloOriginal', 'personagem', 'generoArtista', 'nomeArtista', 'profissao', 'anoNascimento', 'anoFalecimento', 'titulosMaisConhecidos']

# Usando o método drop para remover as colunas
df_filtrado = df_filtrado.drop(columns=colunas_a_remover)
```

```python
# Usando o método drop_duplicates para remover linhas duplicadas
df_filtrado = df_filtrado.drop_duplicates()
```
