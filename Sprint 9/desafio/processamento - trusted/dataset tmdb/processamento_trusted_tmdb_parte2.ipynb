#### Colunas que não são necessárias para a análise foram excluídas. Como os anos de lançamento dos filmes já estão presentes no df "movies.csv" que foi passado pelo PB, a coluna "release_date" deste df também foram excluídas. Dados relacionados aos votos que também exitem no "movies.csv" foram mantidos neste df para análises comparativas entre os dois.

```python
# Removendo colunas
df_tmdb = df_tmdb.drop(['adult','backdrop_path','poster_path','original_title','video','release_date'], axis=1)
```

```python
# Renomeando as colunas
df_tmdb = df_tmdb.rename(columns={'genre_ids': 'id_genero', 'id': 'id_filme','original_language':'idioma_original','overview':'sinopse',
                                  'popularity':'popularidade','title':'titulo','vote_average':'nota_media','vote_count':'qtd_votos'})
```

```python
df_tmdb.head()
```

```python
# Obtendo o número de linhas e colunas
df_tmdb.shape
```

```python
# Removendo os colchetes das linhas na coluna "id_genero"
df_tmdb['id_genero'] = df_tmdb['id_genero'].str.replace(r'\[|\]', '', regex=True)
```
